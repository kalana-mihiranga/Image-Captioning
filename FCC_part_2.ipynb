```python
!pip install -q diffusers transformers accelerate safetensors

# STEP 2: Import Libraries
import torch
from diffusers import StableDiffusionPipeline
from PIL import Image
import matplotlib.pyplot as plt

# STEP 3: Load Pre-trained Stable Diffusion Pipeline
pipe = StableDiffusionPipeline.from_pretrained(
    "runwayml/stable-diffusion-v1-5",
    torch_dtype=torch.float16
).to("cuda")
```

```python
# STEP 4: Define Prompts (You can load from Flickr8k.token.txt later)
prompts = [
    "a little cat playing with a dog in the park",
    "a group of people hiking on a mountain trail",
    "a man riding a red motorcycle on the street",
    "a couple walking hand in hand on the beach",
    "a cat sitting on a window sill looking outside"
]

# STEP 5: Generate Images from Prompts
images = []
for prompt in prompts:
    image = pipe(prompt).images[0]
    images.append(image)

# STEP 6: Display Generated Images
plt.figure(figsize=(20, 10))
for i, img in enumerate(images):
    plt.subplot(1, len(images), i+1)
    plt.imshow(img)
    plt.title(f"Prompt {i+1}")
    plt.axis("off")
plt.tight_layout()
plt.show()

# STEP 7: Save Generated Images (Optional)
for i, img in enumerate(images):
    img.save(f"generated_image_{i+1}.png")

print("✅ Image generation completed.")
```
